Import the packages allowed for this quiz

```python
import numpy as np
import matplotlib.pyplot as plt
```

Create two numpy arrays (X and Y)

np.arange(start, stop) - to generate within the the half open interval

equation for Z takes two values (X,Y)

Z is a Circle - with a center of (5,5)

```python
X = np.arange(0,10)
Y = np.arange(0,10)

X, Y = np.meshgrid(X, Y)

# equation of a circle
Z = (X - 5) ** 2 + (Y - 5) ** 2
```

Plot the contourf figure of Z

```python
plt.figure()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('ContourF graph')
plt.contourf(X, Y, Z,)
plt.colorbar(label='Z')
```

Plot the contour of Z

```python
plt.figure
plt.xlabel('X')
plt.xlabel('Y')
plt.title('contour graph')
plt.contour(Z)
```

Plot the imshow of Z

```python
plt.figure
plt.xlabel('X')
plt.xlabel('Y')
plt.title('imshow graph')
plt.imshow(Z)
```

Plot the pcolormesh of X,Y,Z

```python
plt.figure
plt.xlabel('X')
plt.xlabel('Y')
plt.title('pcolormesh graph')
plt.pcolormesh(X,Y,Z)
```

Quiz Example with experiment Variables

```python
plt.contourf(X,Y,Z, 10, alpha =0.5, cmap = plt.get_cmap('jet'))
plt.plot([5], [5], 'o', ms=12, markeredgewidth=3, color='orange')
plt.xlim(0,10)
plt.ylim(0,10)
plt.xlabel(r'$X$',fontsize=16)
plt.ylabel(r'$Y$',fontsize=16)
plt.colorbar(label='Z')
plt.show()
```
